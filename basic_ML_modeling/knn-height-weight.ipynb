```python
h=[]
w=[]
for i in range(len(ndata)):
    x=ndata.loc[i]["Gender"]
    if x=="Male":
        h.append(ndata.loc[i]["Height"])
        w.append(ndata.loc[i]["Weight"])
```

```python
ndf=pd.DataFrame()
ndf.insert(0,"Height",h)
ndf.insert(1,"Weight",w)
```

```python
X=ndf[["Height","Weight"]]
```

```python
k=4
centroid=(X.sample(n=k))
plt.scatter(X["Height"],X["Weight"],c='red')
plt.scatter(centroid["Height"],centroid["Weight"],c='green')
plt.xlabel("heights")
plt.ylabel("weights")
plt.show()
```

```python
K=4
query=174.77
def knn(K,query,data):
    euclidian_distance=[]
    for i in range(len(data)):
        h=data.loc[i]["Height"]
        w=data.loc[i]["Weight"]
        dis=math.dist([query],[h])
        euclidian_distance.append((dis,w))
#         print(euclidian_distance)
    
    sorted_distance=sorted(euclidian_distance)
    k_neighbours=sorted_distance[:K]
#     print(k_neighbours)
    sum=0
    for i in range(K):
        sum=sum+k_neighbours[i][1]
    mean=sum/k
    return print(mean)
        
        
```

```python
knn(K,query,data)
```
